### Age

```python
average_age_titanic   = data_train["Age"].mean()
average_age_titanic
```

```python
std_age_titanic = data_train["Age"].std()
std_age_titanic
```

```python
count_nan_age_titanic = data_train["Age"].isnull().sum()
count_nan_age_titanic
```

```python
average_age_test   = data_test["Age"].mean()
std_age_test       = data_test["Age"].std()
count_nan_age_test = data_test["Age"].isnull().sum()
```

```python
import numpy as np
rand_1 = np.random.randint(average_age_titanic - std_age_titanic, average_age_titanic + std_age_titanic, size = count_nan_age_titanic)
rand_1
```

```python
rand_2 = np.random.randint(average_age_test - std_age_test, average_age_test + std_age_test, size = count_nan_age_test)
rand_2
```

```python
data_train[data_train["Age"].isnull()]
# data_train.loc[ data_train.Age.isnull()]
```

```python
data_train.loc[ data_train.Age.isnull(),'Age'] = rand_1
```

```python
data_train["Age"] = data_train["Age"].astype(int)
```

```python
data_test.loc[ data_test.Age.isnull(),'Age'] = rand_2
```

```python
data_test["Age"] = data_test["Age"].astype(int)
```

### Cabin

```python
data_train = data_train.drop([ 'Cabin'], axis = 1)
data_test = data_test.drop([ 'Cabin'], axis = 1)
```

```python
data_train
```

### Family

```python
data_train['Family'] =  data_train["Parch"] + data_train["SibSp"]
data_train['Family'].loc[data_train['Family'] > 0] = 1
data_train['Family'].loc[data_train['Family'] == 0] = 0
```

```python
data_test['Family'] =  data_test["Parch"] + data_test["SibSp"]
data_test['Family'].loc[data_test['Family'] > 0] = 1
data_test['Family'].loc[data_test['Family'] == 0] = 0
```

```python
data_train = data_train.drop(['SibSp','Parch'], axis=1)
data_test  = data_test.drop(['SibSp','Parch'], axis=1)
```

### Sex

```python
# data_train.loc[ data_train['Age'] <=16][['Age','Survived']]
data_train[ data_train['Age'] <=16][['Age','Survived']]
```

```python
data_train['Sex'] = data_train['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
```

```python
data_test['Sex'] = data_test['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
```

```python
data_test[data_test["Embarked"].isnull()]
```

```python
data_train['Embarked'] = data_train['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)

data_test['Embarked'] = data_test['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
```

```python
# data_train['Embarked'][data_train['Embarked'] =='S'] = 0
# data_train['Embarked'][data_train['Embarked'] =='Q'] = 1
# data_train['Embarked'][data_train['Embarked'] =='C'] = 2
# data_train['Embarked'] = data_train['Embarked'].astype(int)
```

```python
X_train = data_train.drop("Survived", axis=1)
Y_train = data_train["Survived"]
X_test  = data_test.drop("PassengerId", axis=1).copy()
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = decision_tree.score(X_train, Y_train)
acc_decision_tree
```

```python
submission = pd.DataFrame({
        "PassengerId": data_test["PassengerId"],
        "Survived": Y_pred
    })
submission.to_csv('titanic_pred.csv', index=False)
```
